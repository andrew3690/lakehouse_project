## Initial configs necessarry to run in order to connect spark cluster, hive metastore and postgresql services.

```python
import os
from pyspark.sql import SparkSession

# Pega do ambiente ou usa um default seguro
ms_host = os.getenv('METASTORE_HOST', 'localhost')
ms_user = os.getenv('METASTORE_USER', 'postgres')
ms_pass = os.getenv('METASTORE_PASSWORD', 'hive_password')
```

## Setting spark session for cluster usage

```python
spark = (SparkSession.builder
    .appName("Lakehouse_Clean_Code")
    .master(f"spark://{os.getenv('SPARK_MASTER_HOST')}:7077")
    .enableHiveSupport()
    .getOrCreate())
```
